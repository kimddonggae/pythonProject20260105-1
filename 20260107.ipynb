```python
n=int(input())

isprime=True

for i in range(2, n):
    if n%i ==0:
        print("소수 아님")
        isprime=False
        break

print(isprime)
```

```python
def is_prime(num_list):
    result = []
    for num in num_list:
        isprime = True
        for i in range(2, num):
            if num % i == 0:
                isprime = False
                break
        result.append(isprime)
    return result
f=[1,2,3,4,5]
print(is_prime(f))
```
