```python
f={'홍': '010-2343-2422', '이': '010-2442-2333', '김': '010-2432-0777'}

while True:

    a=int(input("메뉴 선택(1: 검, 2: 추, 0: 종료)"))

    if a==1:
        print(list(f.keys()))
        ad=input('이름')
        if ad in f:
            print(ad, f[ad])
        else :
            print("그친없")
    elif a==2:
        f_key=input("이름")
        f_value=input("전번")
        f[f_key]=f_value
        print(f)
    elif a==0:
        print("종료")
        break
    else: 
        print("오류, 다시 선택")
```

```python
#모스부호 해석
morse = { 

    '.-':'a','-...':'b','-.-.':'c','-..':'d','.':'e','..-.':'f', 

    '--.':'g','....':'h','..':'i','.---':'j','-.-':'k','.-..':'l', 

    '--':'m','-.':'n','---':'o','.--.':'p','--.-':'q','.-.':'r', 

    '...':'s','-':'t','..-':'u','...-':'v','.--':'w','-..-':'x', 

    '-.--':'y','--..':'z' 
} 

a=".--. -.-- - .... --- -."
a_list=a.split()

b=''
for i in a_list:
    print(i)
    b+=morse[i]
    print(morse[i])
print(b)
```
