```python
#미션
사용자로부터 2개 숫자 입력 받아서 덧셈, 뺄셈, 곱셈, 나눗셈 결과 출력해주세요.
```

```python
a=int(input("숫자를 입력하시오:"))
b=int(input("숫자를 입력하시오:"))
print("덧셈:", a, "+", b, "=", a+b)
print("뺄셈:", a, "-", b, "=",a-b)
print("곱셈:", a, "*", b, "=",a*b)
print("나눗셈:", a, "/", b, "=",a/b)
```
